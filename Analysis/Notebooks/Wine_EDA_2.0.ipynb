```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sn
```

```python
file = "../data/winemag-data_first150k.csv"
```

```python
#read in csv 
df = pd.read_csv(file, low_memory=False)
df = df.drop("Unnamed: 0", axis=1)
mask = pd.notnull(df.country)
mask2 = pd.notnull(df.price)
df
```

```python
clean_df = df.loc[mask & mask2].reset_index(drop=True)
clean_df['designation'].replace(np.nan,'*Unknown Wine*', inplace=True)
clean_df.loc[clean_df['region_1'].isnull(),'region_1']=clean_df['province']
clean_df.loc[clean_df['region_2'].isnull(),'region_2']=clean_df['region_1']
clean_df
```

```python
#Load in red wine csv
reds = pd.read_csv('../data/Wine_Reds.csv').dropna()
reds = reds.rename(columns = {"Variety":"variety"})
```

```python
merged_df = pd.merge(clean_df,reds,on = "variety", how = "left")
mask5 = pd.notnull(merged_df["Red?"])
wine_df = merged_df.loc[mask5].reset_index(drop=True)
```

```python
wine_df['wineType_encoded'] = wine_df['Red?'].astype(int)
```

```python
wine_df.drop(columns="Counts",inplace=True)
wine_df
```

```python
wine_df.price.describe()
```

```python
clear_Wine = wine_df.wineType_encoded == 0
clear_Wine = wine_df.loc[clear_Wine]
clear_Wine.price.value_counts()
```

```python
red_Wine = wine_df.wineType_encoded == 1
red_Wine = wine_df.loc[red_Wine]
red_Wine.price.value_counts()
```

```python
price_bins = [0, 10, 15, 20, 30, 50, 2500]
price_labels = ["Value: Under $10", "Popular: $10-15", "Premium: $15-20", "Super Premium: $20-30", "Ultra Premium: $30-50", 
                "Iconic: Over $50"]
wine_df['Wine_Bins'] = pd.cut(wine_df.price, price_bins, labels=price_labels, include_lowest=True)

fullBins = wine_df['Wine_Bins'].copy()

priceCounts = fullBins.value_counts()
perPrice = (priceCounts/len(wine_df))*100
winePrice_table = pd.DataFrame({'Wine Counts':priceCounts, 'Price Range Percent':perPrice.map("{:.0f}%".format)})


winePrice_table
```
